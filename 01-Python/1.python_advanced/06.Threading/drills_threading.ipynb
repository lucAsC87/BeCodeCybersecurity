In the `data/` folder, you have 10 files that contain Shakespears sonnets. You have to gather all these files into one file `data_all.txt` using threads. Be careful, the sonnets must appear in order.

```python
from threading import Thread
import os


class ThreadTest(Thread):
	def __init__(self, text):
		Thread.__init__(self)
		self.text = text

	def run(self):
		with open("data_all.txt", "a") as file:
			file.write(self.text)

start = 1
for file in os.listdir("./data/"):
	name = "data_part_" + str(start) + ".txt"
	file_path = os.path.join("./data/", name)
	with open(file_path, "r") as f:
		thread = ThreadTest(f.read())
		thread.start()
	start += 1
```

### Exercise 2
Scrap all the web pages in the `urls` list and display the links. 1 thread per link. 

```python
urls = [
    "http://www.python.org",
    "http://www.python.org/about/",
    "http://www.onlamp.com/pub/a/python/2003/04/17/metaclasses.html",
    "http://www.python.org/doc/",
    "http://www.python.org/download/",
    "http://www.python.org/getit/",
    "http://www.python.org/community/",
    "https://wiki.python.org/moin/",
]
```

```python
from bs4 import BeautifulSoup
from threading import Thread
import requests

class ThreadLink(Thread):
	def __init__(self, link):
		Thread.__init__(self)
		self.link = link

	def run(self):
		print(self.link['href'])

for url in urls:
	r = requests.get(url)
	soup = BeautifulSoup(r.content, "lxml")
	for p in soup.find_all("a", href=True):
		link_thread = ThreadLink(p)
		link_thread.start()
```
